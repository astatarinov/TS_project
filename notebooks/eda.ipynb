# Balances TS EDA

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from project.utils.data import load_balances_data
```

```python
balances_df = load_balances_data()
```

```python
balances_df.head()
```

```python
balances_df.describe()
```

```python
plt.figure(figsize=(12, 8))

plt.title('Balance Time Series')
plt.xlabel('date')
plt.ylabel('balance')

plt.plot(balances_df['balance'])

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()
plt.grid()
plt.show()
```

```python
print(f'Dickey — Fuller test: p = {sm.tsa.stattools.adfuller(balances_df["balance"])[1]}')
```

```python
sm.tsa.seasonal_decompose(balances_df['balance']).plot()
```
